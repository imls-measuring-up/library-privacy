# Notebook to process the JSON file generated by a library homepage harvest

```python
import json
import csv
```

```python
# define the JSON file to be processed
tablePath = "/Users/kbene/Box Sync/IMLS-Measuring-Up/IR-Privacy/JSON_Files"
harvestTimestampString = "2016-10-05T14_51_11.452293"

sourceJSON = tablePath + "/" + harvestTimestampString + "/result.json"
outFilePath = tablePath + "/" + harvestTimestampString + "/uniqueEndpointFiles.txt"
print sourceJSON
print outFilePath
```

```python
with open(sourceJSON, "r") as sourceFile:
    JSONdict = json.load(sourceFile)
    with open(outFilePath,"w") as outfile:
        headerText = "endURL" + "\t" + "filename" + "\t" + "sourceUUID" + "\n"
        outfile.write(headerText)
        
        uniqueURLs = []
        for itemKey in JSONdict.iterkeys():
            outText = JSONdict[itemKey]["pageRedirectEndURL"] + "\t" + JSONdict[itemKey]["filename"] + "\t" + JSONdict[itemKey]["sourceUUID"] + "\n"
            # compare the endURL of the current item with the list of items that have already
            # been written. To remove URLs that are the same except for a trailing "/" the
            # strip() function was added into the comparison and writing of URLs into the list
            if JSONdict[itemKey]["pageRedirectEndURL"].strip('/') not in uniqueURLs:
                outfile.write(outText.encode('utf8'))
                print "adding " + JSONdict[itemKey]["pageRedirectEndURL"]
                uniqueURLs.append(JSONdict[itemKey]["pageRedirectEndURL"].strip('/'))
            else:
                print "skipping " + JSONdict[itemKey]["pageRedirectEndURL"]
```
